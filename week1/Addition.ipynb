```python
from model import Model
from dmchunk import Chunk
import pandas as pd
```

```python
df = pd.DataFrame(columns = ["subject","trial","RT"])
```

```python
def addition(m, num1, num2):
    g = Chunk(name = "goal", slots = {"isa": "addition-goal", "num1":num1, "num2":num2})
    m.goal = g
    done = False
    while not done:
        if not "sum" in g.slots:
            m.time += .05
            g.slots["sum"] = g.slots["num1"]
            g.slots["count"] = "zero"
        elif g.slots["count"] != g.slots["num2"]:
            m.time += 0.05
            request = Chunk(name = "request", slots = {"isa":"count-fact","num1":g.slots["count"]})
            chunk, latency = m.retrieve(request)
            m.add_encounter(chunk)
            m.time += latency
            g.slots["count"] = chunk.slots["num2"]
            m.time += 0.05
            request = Chunk(name = "request", slots = {"isa":"count-fact","num1":g.slots["sum"]})
            chunk, latency = m.retrieve(request)
            m.add_encounter(chunk)
            m.time += latency
            g.slots["sum"] = chunk.slots["num2"]
            print(g.slots["sum"])
            m.time += 0.3
        else:
            m.time += 0.05
            m.time += 0.3
            print(m.time)
            print(g.slots["sum"])
            done = True
       
```

```python
n = 100
k = 10
for subj in range (0,n):
    m  = Model()
    numbers = ["zero","one","two","three","four","five","six","seven","eight","nine"]
    for i in range(0,len(numbers)-1):
        fact = Chunk(name = "cf" + numbers[i], slots = {"isa":"count-fact", "num1": numbers[i], "num2" : numbers[i+1]})
        m.add_encounter(fact)
    for trial in range(0,k):
        start_time = m.time
        addition(m, "two", "three")
        df.loc[subj * k + trial] = [subj, trial, m.time - start_time]
```
